```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# importing the companies csv file
companies = pd.read_csv('companies.txt', sep = '\t', encoding = 'ISO-8859-1')
```

```python
#inspecting companies csv file
companies
```

```python
#cleaning the imported data for companies data frame
companies['permalink'] = companies['permalink'].str.encode('utf-8').str.decode('ascii', 'ignore').str.lower()
companies['name'] = companies['name'].str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
#finding unique count of companies in companies csv
len(companies['permalink'].unique())
```

```python
# importing rounds2 csv
rounds2 = pd.read_csv('rounds2.csv', encoding = 'ISO-8859-1')
```

```python
#inspecting csv
rounds2
```

```python
# cleaning the data from rounds2
rounds2['company_permalink'] = rounds2['company_permalink'].str.encode('utf_8').str.decode('ascii', 'ignore').str.lower()
```

```python
# inspecting rounds2 dataframe
rounds2
```

```python
# finding unique count of round2
len(rounds2['company_permalink'].unique())
```

```python
# verifying the list of entities that are there in rounds2 but not in companies
print(len(rounds2.loc[~rounds2['company_permalink'].isin(companies['permalink']), :]))
```

```python
# creating a master frame
master_frame = pd.merge(companies, rounds2, how = 'inner', left_on = 'permalink', right_on = 'company_permalink')
```

```python
master_frame
```

```python
# inscpecting null values 
round(master_frame.isnull().sum()/len(master_frame.index)*100,2)
```

```python
#shape before cleaning
master_frame.shape
```

```python
# Dropping null values from raised_amount_usd column
master_frame.drop(master_frame[master_frame['raised_amount_usd'].isnull()].index, inplace = True)
```

```python
#shape after cleaning
master_frame.shape
```

```python
# dropping values with 0 investment (as it would affect our average calulations and no investment would be of 0 value)
master_frame.drop(master_frame[master_frame['raised_amount_usd'] == 0].index, inplace = True)
```

```python
#shape after deleting 0 values
master_frame.shape
```

```python
#inspecting the master frame
master_frame
```

```python
#to avoid scientific notation
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# Average investment - venture
master_frame.groupby('funding_round_type').mean().loc[['venture']]
```

```python
# Average investment - angel
master_frame.groupby('funding_round_type').mean().loc[['angel']]
```

```python
# Average investment - seed
master_frame.groupby('funding_round_type').mean().loc[['seed']]
```

```python
# Average investment - private equity
master_frame.groupby('funding_round_type').mean().loc[['private_equity']]
```

```python
# avg investment amount - computed together for all 4 types
pd.DataFrame(master_frame.groupby('funding_round_type').raised_amount_usd.mean()).loc[['venture', 'angel', 'seed', 'private_equity']].sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
#Avg investment in our invetment range - as Spark Funds would be interested to find the average investments within their investment range for each type of funding
#And as per question one with the best average has to be selected, hence private equity emerges as the best option
master_frame[(master_frame['raised_amount_usd'] >= 5000000) & 
             (master_frame['raised_amount_usd'] <= 15000000)].groupby('funding_round_type').mean().loc[['angel', 'seed', 'private_equity', 'venture']].sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
#creating top9 data frame
top9 = pd.DataFrame(master_frame[master_frame['funding_round_type'] == 'private_equity'].groupby('country_code').
             raised_amount_usd.sum()).sort_values(by = 'raised_amount_usd', ascending = False)[:9]
```

```python
#inspecting top9 dataframe
top9
```

```python
#creating a list of country for filtering
country_list = list(top9.index)
```

```python
#creating a DF of top 9 countries and funding type private equity
top9_df = master_frame[(master_frame['country_code'].isin(country_list)) & (master_frame['funding_round_type'] == 'private_equity')]
```

```python
#inspecting shape
top9_df.shape
```

```python
#Inspecting Data frame
top9_df
```

```python
#splitting the category list column into primary sector
top9_df['primary_sector'] = top9_df['category_list'].str.split('|', n = 2, expand = True)[[0]]
top9_df
```

```python
#inspecting top9 df
top9_df
```

```python
#importing mapping csv
mapping = pd.read_csv('mapping.csv')
```

```python
#inspecting mapping
mapping
```

```python
# dropping first row
mapping.drop(0, inplace = True)
```

```python
#melting mapping csv
mapping = pd.melt(mapping, id_vars=['category_list'])
```

```python
#inspecting mapping
mapping
```

```python
# filtering out relevant rows
mapping = mapping[mapping['value'] ==1]
```

```python
#dropping additional column
mapping.drop('value', axis = 1, inplace = True)
```

```python
#renaming columns
mapping.rename(columns={'category_list':'primary_sector', 'variable':'main_sector'}, inplace = True)
```

```python
#inspecting mapping
mapping
```

```python
# merging maping and top9 df
#using left join to retain all records of top9_df
top9_df = pd.merge(top9_df, mapping, how = 'left', on = 'primary_sector')
```

```python
top9_df
```

```python
#inspecting main sector with null values
top9_df[top9_df['main_sector'].isnull()]
```

```python
#imputing values for null values in main sector
# setting primary sector for analytics as is a major_sector for it
top9_df.loc[top9_df['primary_sector'].str.contains('Analytics', na = False),['main_sector']] = 'Social, Finance, Analytics, Advertising'
```

```python
#imputing values for null values in main sector
# setting primary sector for finance as there is a major sector for it
top9_df.loc[top9_df['primary_sector'].str.contains('Financ', na = False),['main_sector']] = 'Social, Finance, Analytics, Advertising'
```

```python
#imputing for null values
#setting value as others
top9_df.loc[top9_df['main_sector'].isnull(), ['main_sector']] = 'Others'
```

```python
#inspecting balance null values
len(top9_df[top9_df['main_sector'].isnull()])
```

```python
# creating d1 USA
d1 = top9_df[(top9_df['country_code'] == 'USA') & ((top9_df['raised_amount_usd'] >= 5000000) & (top9_df['raised_amount_usd'] <= 15000000))]
```

```python
#creating d1 sector_analysis
d1_sector_analysis = d1.groupby('main_sector')['raised_amount_usd'].agg(('sum', 'count'))
```

```python
# remaing columns
d1_sector_analysis.rename(columns={'sum':'Sum of investment', 'count': 'Count of Investment'}, inplace = True)
```

```python
#merging count and investment amount
d1 = pd.merge(d1, d1_sector_analysis, on = 'main_sector')
```

```python
#inspecting d1
d1
```

```python
#Creating d2 India
d2 = top9_df[(top9_df['country_code'] == 'IND') & ((top9_df['raised_amount_usd'] >= 5000000) & (top9_df['raised_amount_usd'] <= 15000000))]
```

```python
# creating sector analysis for d2
d2_sector_analysis = d2.groupby('main_sector')['raised_amount_usd'].agg(('sum', 'count'))
```

```python
#renaming columns
d2_sector_analysis.rename(columns={'sum':'Sum of investment', 'count': 'Count of Investment'}, inplace = True)
```

```python
#merging count and investment amount
d2 = pd.merge(d2, d2_sector_analysis, on = 'main_sector')
```

```python
#inspecting
d2
```

```python
#Creating d3 GBR
d3 = top9_df[(top9_df['country_code'] == 'GBR') & ((top9_df['raised_amount_usd'] >= 5000000) & (top9_df['raised_amount_usd'] <= 15000000))]
```

```python
d3
```

```python
# creating sector analysis for d3
d3_sector_analysis = d3.groupby('main_sector')['raised_amount_usd'].agg(('sum', 'count'))
```

```python
#renaming columns
d3_sector_analysis.rename(columns={'sum':'Sum of investment', 'count': 'Count of Investment'}, inplace = True)
```

```python
#merging count and investment amount
d3 = pd.merge(d3, d3_sector_analysis, on = 'main_sector')
```

```python
d3
```

```python
# concatinating d1, d2 and d3
top3 = pd.concat([d1, d2, d3], axis = 0, ignore_index = True, sort = False)
```

```python
#inspecting
top3
```

```python
# count of investment
top3.groupby('country_code')['raised_amount_usd'].count()
```

```python
# sum of investment
top3.groupby('country_code')['raised_amount_usd'].sum()
```
